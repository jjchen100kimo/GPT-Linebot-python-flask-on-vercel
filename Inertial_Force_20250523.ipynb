```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Parameters
F = 1.0  # Amplitude
t = np.linspace(0, np.pi, 1000)  # Time array from 0 to 
#omega_values = [0.5, 1.0, 2.0, 4.0]  # Different omega values
omega_values = np.linspace(0, 5, 20)  # Modified: from 0 to 10, divided into 100 parts

# Create a figure with two subplots: one for the sine waves, one for integrals
plt.figure(figsize=(12, 8))

# Plotting sine waves
plt.subplot(2, 1, 1)
for omega in omega_values:
    y = F * np.sin(omega * t)
    plt.plot(t, y, label=f' = {omega}')
plt.title('Sine Wave: F*sin(t)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Calculating and plotting the integral for each omega
integrals = []
for omega in omega_values:
    # Define the function to integrate: F*sin(t)
    func = lambda t: F * np.sin(omega * t)
    # Compute the definite integral from 0 to 2
    # integral, _ = integrate.quad(func, 0, 2*np.pi)
    integral, _ = integrate.quad(func, 0, np.pi/omega)
    integrals.append(integral)

# Plotting the integrals
plt.subplot(2, 1, 2)
plt.bar([str(omega) for omega in omega_values], integrals, color='skyblue')
plt.title('Integral of F*sin(t) from 0 to /omega ')
plt.xlabel('')
plt.ylabel('Integral Value')
plt.grid(True)

# Adjust layout and save the plot
plt.tight_layout()
# plt.savefig('sine_wave_analysis.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Parameters
F = 1.0  # Amplitude
t = np.linspace(0, 10, 1000)  # Time array from 0 to 2
omega_values = np.linspace(0, 10, 100)  # From 0 to 10, divided into 100 parts


# Create a figure with two subplots: one for the sine waves, one for integrals
plt.figure(figsize=(12, 8))

# Plotting sine waves (only every 10th omega to reduce clutter)
plt.subplot(2, 1, 1)
for omega in omega_values[::10]:  # Select every 10th value
    y = F * np.sin(omega * t)
    plt.plot(t, y, label=f' = {omega:.3f}')  # Format to 3 decimal places
plt.title('Sine Wave: F*sin(t)')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Calculating and plotting the integral for each omega
integrals = []
for omega in omega_values:
    # Define the function to integrate: F*sin(t)
    func = lambda t: F * np.sin(omega * t)
    # Compute the definite integral from 0 to 
    integral, _ = integrate.quad(func, 0, np.pi/omega)
    integrals.append(round(integral, 3))  # Round integral to 3 decimal places

# Plotting the integrals
plt.subplot(2, 1, 2)
plt.bar([f'{omega:.3f}' for omega in omega_values], integrals, color='skyblue')  # Format to 3 decimal places
plt.title('Integral respect with t -- F*sin(t) from 0 to ')
plt.xlabel('')
plt.ylabel('Integral Value')
plt.grid(True)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Limit number of x-ticks

# Adjust layout and save the plot
plt.tight_layout()
# plt.savefig('sine_wave_analysis.png')
```
